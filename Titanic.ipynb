```python
# Converting categorical columns into numerical by importing LabelEncoder from preprocessing module of sklearn library
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Name'] = le.fit_transform(df['Name'])
df['Name']
df['Sex'] = le.fit_transform(df['Sex'])
df['Sex']
df['Age'] = le.fit_transform(df['Age'])
df['Age']
df['Ticket'] = le.fit_transform(df['Ticket'])
df['Ticket'] 
df['Cabin'] = le.fit_transform(df['Cabin'])
df['Cabin']
df['Embarked'] = le.fit_transform(df['Embarked'])
df['Embarked']
```

```python
# Using train_test_split model to predict values 
X = df.drop(columns=['Sex','Survived'])
Y = df['Sex']

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.10)

X_test.shape
```

```python
X_train = X_train.dropna()
Y_train = Y_train.loc[X_train.index]
```

```python
#Importing linear model to create model and predict values
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train) # Giving values to the model
```

```python
#Import accuracy score from metrics to check the accuracy of the prediction made by the model.
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_train)
score = accuracy_score(Y_train, y_pred)
accuracy = score*100
print(accuracy)
```
